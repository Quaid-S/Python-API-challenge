```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os


from api_keys import g_key

```

```python
path = "weather.csv"
df = pd.read_csv(path)
df.head()
```

```python
from api_keys import g_key
gmaps.configure(api_key=g_key)
```

```python
df = df.dropna()
coords = df[["Lat", "Lng"]]
weight = df["Humidity"]
```

```python
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(coords, weights=weight, dissipating=False, point_radius=5)
fig.add_layer(heat_layer)
fig
```

```python
#A max temperature lower than 80 degrees but higher than 70.

df_temp = df.loc[df['Max Temp'] < 26.67 ]
df_temp= df_temp.loc[df['Max Temp'] >  21.11]
df_temp.head(30)
```

```python
#Wind speed less than 10 mph.

df_wind = df_temp.loc[df_temp['Wind Speed'] < 10]
```

```python
#Zero cloudiness
df_cloud = df_wind.loc[df_wind['Cloudiness'] == 0.0]
```

```python
#Using Google Places API to find the first hotel for each city located within
#5000 meters of your coordinates.

df_final = df_cloud

coords = df_final[['Lat', 'Lng']]
target_type = 'lodging'
target_radius = 5000
target_keyword = 'Hotel'

params = {
    "location": coords,
    "keyword": target_keyword,
    "radius": target_radius,
    "type": target_type,
    "key": g_key
}
```

```python
query_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
Hotels = []
for coord in coords:
    response = requests.get(query_url, params=params).json()
    Hotels.append(response)
    
```
